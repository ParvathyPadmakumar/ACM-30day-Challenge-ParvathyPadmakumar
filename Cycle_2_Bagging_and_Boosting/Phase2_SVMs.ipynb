```python
import pandas as pd
import numpy as np

df=pd.read_csv('creditcard.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
x=df.drop(['Time','Class'],axis=1)
y=df['Class']
scaler=StandardScaler()
x=scaler.fit_transform(x)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import LinearSVC
print("Before PCA")
classifier=LinearSVC(C=0.1,random_state=42)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print('Accuracy of linear model',accuracy)

classifier=SVC(kernel='rbf',C=0.1,gamma='scale',random_state=42)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print('Accuracy of RBF model',accuracy)

classifier=SVC(kernel='poly',C=0.1,random_state=42,gamma='scale',degree=3)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print('Accuracy of polynomial model',accuracy)
```

The above code took 3m27s to run . Hence,implementing PCA as the dataset is huge.

```python
print("setting c value to 0.1")
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10,random_state=42)
x=pca.fit_transform(x)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
classifier=SVC(kernel='linear',C=0.1, random_state=42)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print('Accuracy of linear SVM after PCA',accuracy)

classifier=SVC(kernel='rbf',C=0.1,gamma='scale',random_state=42)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print('Accuracy of RBF SVM after PCA',accuracy)

classifier=SVC(kernel='poly',C=0.1,random_state=42,gamma='scale',degree=2)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print('Accuracy of polynomial SVM after PCA',accuracy)
```

```python
pca=PCA(n_components=2,random_state=42)
xpca=pca.fit_transform(x)
plt.figure(figsize=(10, 6))
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA plot')
plt.scatter(xpca[:,0],xpca[:,1],c=y,alpha=0.5)
plt.colorbar(label='Class')
plt.show()

```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x=xpca[:,0], y=xpca[:,1],hue=y,alpha=0.5)
plt.title('2D PCA plot-seaborn')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Class')
plt.show()
```
