```python
import pandas as pd
import numpy as np

df=pd.read_csv('Iris.csv')
df.head()
```

```python

df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

x=df.drop(['Id','Species'],axis=1)

tsne=TSNE(n_components=2,random_state=42)
xtsne=tsne.fit_transform(x)
print("T-sne implemented for dim reduction")
tsne.kl_divergence_
```

```python
print("KMeans clustering")
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(x)
ypredict=kmeans.predict(x)
kmeanslabels=ypredict
plt.figure(figsize=(10,3))
plt.scatter(xtsne[:,0],xtsne[:,1],c=kmeanslabels,cmap='viridis')
plt.title('KMeans Clustering t-SNE')
plt.xlabel('t-SNE component 1')
plt.ylabel('t-SNE component 2')
plt.show()
```

```python
from scipy.cluster import hierarchy
plt.figure(figsize=(10,4))
dendrogram=hierarchy.dendrogram(hierarchy.linkage(xtsne,method='ward'))
plt.title('Hierarchical clustering(dendrogram)')
plt.ylabel("Euclidean distances")
plt.xlabel("data points")
plt.show()
```
